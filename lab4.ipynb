```python
# Вправа 2 (1)

import numpy as np
import time

def task_2_numpy():
    n = int(input("n = "))
    start = time.time()
    arr = np.zeros((n, n), dtype=int)
    arr[::2, 1::2] = 1
    arr[1::2, ::2] = 1
    end = time.time()
    print(arr)
    print("Час виконання NumPy:", end - start)
```

$$
\begin{cases}
0\cdot x_1 + 1\cdot x_2 - 3\cdot x_3 + 4\cdot x_4 = -5\\[4pt]
1\cdot x_1 + 0\cdot x_2 - 2\cdot x_3 + 3\cdot x_4 = -4\\[4pt]
1\cdot x_1 + 1\cdot x_2 + 2\cdot x_3 - 5\cdot x_4 = 12\\[4pt]
4\cdot x_1 + 3\cdot x_2 - 5\cdot x_3 + 0\cdot x_4 = 5
\end{cases}
$$

<p style="text-align: center;">або, у матричному вигляді:</p>

$$
A=\begin{pmatrix}
0 & 1 & -3 & 4\\
1 & 0 & -2 & 3\\
1 & 1 & 2 & -5\\
4 & 3 & -5 & 0
\end{pmatrix},\qquad
\mathbf{b}=\begin{pmatrix}-5\\-4\\12\\5\end{pmatrix}.
$$

```python
import numpy as np

A = np.array([[0, 1, -3, 4],
              [1, 0, -2, 3],
              [1, 1,  2, -5],
              [4, 3, -5, 0]], dtype=float)
b = np.array([-5, -4, 12, 5], dtype=float)

# Метод Крамера
detA = np.linalg.det(A)
if abs(detA) < 1e-12:
    raise ValueError("det(A) ~ 0, а отже метод Крамера непридатний.")
x_cramer = np.zeros(4)
for i in range(4):
    Ai = A.copy()
    Ai[:, i] = b
    x_cramer[i] = np.linalg.det(Ai) / detA
print("Розв’язок:", x_cramer)

# Перевірка a) A @ x
Ax = A @ x_cramer
print("A @ x_cramer =", Ax)
print("Співпадає з b:", np.allclose(Ax, b))

# Перевірка b) A^{-1} * b
x_inv = np.linalg.inv(A) @ b
print("\nA^{-1} * b =", x_inv)

# Перевірка c) numpy.linalg.solve()
x_solve = np.linalg.solve(A, b)
print("np.linalg.solve(A, b) =", x_solve)

# Порівняння
print("\nПорівняння результатів (allclose):")
print("Крамер vs A^{-1}b -", np.allclose(x_cramer, x_inv))
print("Крамер vs np.linalg.solve() -", np.allclose(x_cramer, x_solve))
print("A^{-1}b vs np.linalg.solve() -", np.allclose(x_inv, x_solve))
```

$$
(A^2 - B^2)(A + B),
$$

<p style="text-align: center;">де</p>

$$
A = \begin{pmatrix}
7 & 2 & 0 \\
-7 & -2 & 1 \\
1 & 1 & 0
\end{pmatrix}, \qquad
B = \begin{pmatrix}
0 & 2 & 3 \\
1 & 0 & -2 \\
3 & 1 & 1
\end{pmatrix}.
$$

```python
# NumPy

import numpy as np
import time

A = np.array([[7, 2, 0],
              [-7, -2, 1],
              [1, 1, 0]], dtype=float)
B = np.array([[0, 2, 3],
              [1, 0, -2],
              [3, 1, 1]], dtype=float)
start = time.time()
result_numpy = (A @ A - B @ B) @ (A + B)
end = time.time()
print("Результат (NumPy):")
print(result_numpy)
print("Час виконання NumPy:", end - start)
```

```python
# Цикли

A_list = [[7, 2, 0],
          [-7, -2, 1],
          [1, 1, 0]]
B_list = [[0, 2, 3],
          [1, 0, -2],
          [3, 1, 1]]

def matmul(X, Y):
    return [[sum(X[i][k] * Y[k][j] for k in range(len(Y))) for j in range(len(Y[0]))] for i in range(len(X))]
    
def matadd(X, Y):
    return [[X[i][j] + Y[i][j] for j in range(len(X[0]))] for i in range(len(X))]

def matsub(X, Y):
    return [[X[i][j] - Y[i][j] for j in range(len(X[0]))] for i in range(len(X))]

# (A^2 - B^2)(A + B)
start = time.time()
A2 = matmul(A_list, A_list)
B2 = matmul(B_list, B_list)
diff = matsub(A2, B2)
sumAB = matadd(A_list, B_list)
result_loop = matmul(diff, sumAB)
end = time.time()
print("Результат (цикли):")
for row in result_loop:
    print(row)
print("Час виконання циклів:", end - start)
```
